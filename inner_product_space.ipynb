```
import torch as t
import numpy as np
```

```
def inner_product(x,y,A):
  "return transformed inner product"
  return t.matmul(t.matmul(x.t(), A),y)

def norm (x,A):
  "return transformed norm of vector"
  return t.sqrt(t.matmul(x.t(),t.matmul(A,x)))

def angle(x,y,A):
  'returns tuple of angle between x,y in (rad,deg)'
  numerator =  inner_product(x,y,A)
  denumerator =  norm(x, A) * norm(y, A)
  rad = t.acos(numerator/denumerator)
  deg = t.rad2deg(rad) 
  return rad,deg
  
```

```
x = t.tensor([1.,1.])
y = t.tensor([-1.,1.])
T = t.tensor([[2.,-1.],[-1.,4.]])
```

```
angle(x,y,T)
```
